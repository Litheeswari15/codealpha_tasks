# **Import Libraries**

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# **Load Dataset**

```python
df = pd.read_csv('employee_wellbeing_productivity.csv')

sns.set(style='whitegrid')
```

# **Plot histograms**

```python
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
sns.histplot(df['work_hours_per_day'], bins=20, kde=True, color='skyblue')
plt.title('Work Hours Per Day Distribution')

plt.subplot(1, 2, 2)
sns.histplot(df['productivity_score'], bins=20, kde=True, color='salmon')
plt.title('Productivity Score Distribution')

plt.tight_layout()
plt.show()

```

# ** Correlation heatmap**

```python
plt.figure(figsize=(10, 8))
numeric_cols = df.select_dtypes(include=['float64', 'int64']).columns
sns.heatmap(df[numeric_cols].corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap')
plt.show()
```

# **Scatterplot**

```python
plt.figure(figsize=(7, 5))
sns.scatterplot(
    x='work_hours_per_day', 
    y='productivity_score', 
    hue='gender', 
    data=df
)
plt.title('Work Hours vs Productivity Score')
plt.show()
```

# **Boxplot for remote work**


```python
plt.figure(figsize=(6, 5))
sns.boxplot(
    x='remote_work', 
    y='productivity_score', 
    data=df
)
plt.title('Productivity Score by Remote Work Status')
plt.show()
```

# **Barplot for diet quality**

```python
plt.figure(figsize=(7, 5))
sns.barplot(
    x='diet_quality', 
    y='productivity_score', 
    data=df, 
    estimator='mean', 
    ci='sd'
)
plt.title('Average Productivity Score by Diet Quality')
plt.show()
```

# **Pairplot for selected variables**

```python
sns.pairplot(
    df[['work_hours_per_day', 'sleep_hours', 'exercise_frequency', 'productivity_score']],
    kind='scatter'
)
plt.suptitle('Pairwise Relationships', y=1.02)
plt.show()
```
